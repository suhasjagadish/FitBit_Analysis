```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv("FitBit data.csv")
```

Data is copied to another refernce for further analysis so that the original data is not affected by the changes made during analysis.

```python
data_copy.TotalDistance.nunique()
```

```python
data_copy.TrackerDistance.nunique()
```

```python
compare = data_copy["TotalDistance"] == data_copy["TrackerDistance"]
```

```python
compare.value_counts()
```

```python
data_copy[compare == False]
```

1. Since we see that there could be some error in the TrackerDistance and is almost similar to the TotalDistance, so we can drop TrackerDistance. 
2. Since ID column is just a unique number for the activity it does not help in the analysis so we can drop it. 

```python
data_copy.drop(columns = ["TrackerDistance", "Id"],inplace = True)
```

```python
data_copy.head()
```

```python
data_copy.ActivityDate.dtype
```

Since ActivityDate is in object format we need to convert it into a DateTime Format for the ease of analysis.

```python
data_copy["ActivityDate"] = (pd.to_datetime(data_copy["ActivityDate"]))
```

```python
data_copy.head()
```

```python
data_copy.ActivityDate.dtype
```

```python
data_copy.ActivityDate.unique()
```

```python
data_copy["Year"] = pd.DatetimeIndex(data_copy.ActivityDate).year
```

```python
data_copy["Month"] = pd.DatetimeIndex(data_copy.ActivityDate).month
```

```python
data_copy["Date"] = pd.DatetimeIndex(data_copy.ActivityDate).day
```

```python
data_copy.head(2)
```

```python
data_copy.Month.value_counts()
```

```python
data_copy = data_copy.sort_values(["Month","Date"])
```

```python
data_copy.head(10)
```

```python
plt.figure(figsize = (12,8))
sns.set(font_scale = 1.5)
ax = sns.boxplot(x='Date', y='Calories', data=data_copy, color = "pink")
sns.stripplot(data = data_copy, x = "Date", y = "Calories", jitter = False, dodge = True,color = "grey")
plt.title("Date vs Calories")
plt.show()
```

```python
data_copy["Week"] = pd.DatetimeIndex(data_copy.ActivityDate).week
```

```python
data_copy.head(2)
```

```python
data_copy["Day"] = data_copy['ActivityDate'].dt.day_name()
```

```python
data_copy
```

```python
data_copy.Year.value_counts()
```

```python
plt.hist(data_copy["Year"])
```

This confirms that this data is only of 2016 so we can drop this coloumn 

```python
data_copy.drop(columns = "Year", inplace = True)
```

```python
data_copy.head()
```

```python
sns.distplot(data_copy["Date"],bins = None);
```

This shows that the users are very active at the start of the month and little active at mid and moderately active at end of a month

```python
plt.figure(figsize = (14,7))
sns.set(font_scale = 1.5)
sns.histplot(x = data_copy["Day"])
plt.show()
```

This shows that the users are active at Weekend and Weekstart and less active at mid of the week

```python
data_copy.Day.value_counts()
```

```python
plt.figure(figsize = (12,8))
sns.set(font_scale = 1.5)
sns.swarmplot(data = data_copy, x = "Day", y = "Calories", dodge = True)
plt.show()
```
